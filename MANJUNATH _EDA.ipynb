```python
sns.countplot(data=customers, x='Region')
plt.title("Customer Distribution by Region")
plt.show()
plt.savefig('Customer Distribution by Region.png')
```

```python
top_products = transactions.groupby('ProductID')['Quantity'].sum().nlargest(5)
top_products.plot(kind='bar', title="Top 5 Products by Sales")
plt.show()
```

```python
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
revenue_trend = transactions.groupby(transactions['TransactionDate'].dt.date)['TotalValue'].sum()
revenue_trend.plot(title="Revenue Over Time")
plt.show()
```

```python
customer_spend = transactions.groupby('CustomerID')['TotalValue'].sum()
customer_spend.plot(kind='hist', bins=20, title="Customer Spending Distribution")
plt.show()
```

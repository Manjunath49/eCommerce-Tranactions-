```python
merged = transactions.merge(customers, on="CustomerID").merge(products, on="ProductID")
merged
```

```python
customer_features = merged.groupby("CustomerID").agg({
    "TotalValue": "sum",      # Total spending
    "Quantity": "sum",        # Total quantity purchased
    "ProductID": "nunique"    # Number of unique products purchased
}).rename(columns={"ProductID": "ProductDiversity"}).reset_index()
customer_features
```

```python
from sklearn.preprocessing import MinMaxScaler

numeric_columns = customer_features.select_dtypes(include=["int64", "float64"]).columns
scaler = MinMaxScaler()
customer_features_scaled = customer_features.copy()
customer_features_scaled[numeric_columns] = scaler.fit_transform(customer_features[numeric_columns])
```

```python
customer_features["TotalValue"] = pd.to_numeric(customer_features["TotalValue"], errors="coerce")
customer_features["Quantity"] = pd.to_numeric(customer_features["Quantity"], errors="coerce")
```

```python
customer_features.fillna(0, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
numeric_columns = customer_features.select_dtypes(include=["int64", "float64"]).columns
scaler = MinMaxScaler()
customer_features_scaled = customer_features.copy()
customer_features_scaled[numeric_columns] = scaler.fit_transform(customer_features[numeric_columns])
```

```python
print(customer_features_scaled.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(customer_features_scaled.iloc[:, 1:])
```

```python

top_3_similar = {}

for i, cust_id in enumerate(customer_features_scaled["CustomerID"][:20]):
    similarities = list(enumerate(similarity_matrix[i]))
 
    similarities = sorted(similarities, key=lambda x: x[1], reverse=True)[1:4]

    top_3_similar[cust_id] = [(customer_features_scaled["CustomerID"][j], round(score, 2)) for j, score in similarities]
```

```python
top_3_similar
```

```python
import csv

with open("Lookalike.csv", "w") as file:
    writer = csv.writer(file)
    writer.writerow(["cust_id", "list<cust_id, score>"])
    for cust_id, similar in top_3_similar.items():
        writer.writerow([cust_id, similar])
```
